```python
from random import randint
class Dice():
    def __init__(self,sides=6):
        self.sides=sides
    def roll(self):
        return randint(1,self.sides)
```

```python
Dice1=Dice()
results=[]
for i in range(10):
    result=Dice1.roll()
    results.append(result)
frequencies=[]
for i in range(1,Dice1.sides+1):
    frequency=results.count(i)
    frequencies.append(frequency)
print(results)
print(frequencies)
```

```python
import pygal
bar_graph=pygal.Bar()
bar_graph.add('Sai',frequencies)
bar_graph.render_to_file("bar.svg")
```
