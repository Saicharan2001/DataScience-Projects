```python
#Supervised learning - Regression model - LinearRegression algorithm

#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing dataset from remote link
url="https://bit.ly/w-data"
s_data=pd.read_csv(url)
print("data imported successfully")
s_data.head(10)
```

```python
s_data.plot(x='Hours',y='Scores')
plt.title("Hours vs Percentage")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
s_data.plot(x='Hours',y='Scores',style='o')
plt.title("Hours vs Percentage")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
#Preparing the data
#Dividing the data into inputs(attributes) and outputs(labels)
x=s_data.iloc[:,:-1].values
y=s_data.iloc[:,1].values
#above is the first way and below the second way
#x=s_data[['Hours']].values
#y=s_data[['Scores']].values
```

```python
y,x
```

#now we need to spilt the data into training and test data sets by using Scikit_Learns built in method train_test_split() method:

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

#Now we need to train the algorithm
from sklearn.linear_model import LinearRegression
Object_regressor=LinearRegression()
Object_regressor.fit(x_train,y_train)
print("Training is completed")
```

```python
#plotting the regression line
line=Object_regressor.coef_*x+Object_regressor.intercept_   #( It is of form y=mx+c)
plt.scatter(x,y)
plt.plot(x,line)
plt.title("Comparison of study hours with results")
plt.show()
```

```python
#Making predictions
print(x_test)
y_pred=Object_regressor.predict(x_test)
print(y_pred)
print(y_test)
```

```python
#Now getting the prediction
score_pred=np.array([9.25])
score_pred=score_pred.reshape(-1,1)
predict=Object_regressor.predict(score_pred)
print("No of hours:",9.25)
print("Result predicted:",predict[0])
```

```python
#Final step : Evaluating the linear regression model
from sklearn import metrics
print("Mean absolute error:",metrics.mean_absolute_error(y_test,y_pred))
```
